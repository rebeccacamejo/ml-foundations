## Convolutional Filters: A Handsâ€‘On Demonstration

Convolutional Neural Networks (CNNs) use **convolution** operations to extract
local patterns from images.  Instead of training a full CNN here, we
illustrate how simple filters such as **edge detectors** can reveal structure
within an image.  This intuition explains why CNNs excel at tasks like
image recognition.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn.datasets import load_digits

%matplotlib inline

# Load a sample digit
digits = load_digits()
image = digits.images[0]

# Define simple convolution filters
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

# Apply filters
edge_x = ndimage.convolve(image, sobel_x)
edge_y = ndimage.convolve(image, sobel_y)
edges = np.hypot(edge_x, edge_y)

# Plot original and filtered images
fig, axes = plt.subplots(1, 3, figsize=(9, 3))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')
axes[0].axis('off')
axes[1].imshow(edge_x, cmap='gray')
axes[1].set_title('Sobel X')
axes[1].axis('off')
axes[2].imshow(edges, cmap='gray')
axes[2].set_title('Edge Magnitude')
axes[2].axis('off')
plt.suptitle('Demonstrating Convolutional Edge Detection')
plt.show()
```

### Interpretation

The simple Sobel filters detect horizontal and vertical edges in the digit.  When combined, they reveal the full edge structure.  **Convolutional neural networks** learn such filters automatically and stack them to recognise increasingly abstract features (e.g., corners, textures, shapes).  This example demonstrates how convolution captures local patterns essential for image understanding.  Businesses use similar techniques for quality control (detecting defects) or automated document processing.
